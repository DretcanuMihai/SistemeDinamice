```python
#c)
forget()
var("omega_0,t,x_0,v_0")
x=function("x")(t)
deq=diff(x,t,2)+(omega_0^2)*x==0 #w_0=g/l
##show(deq)
assume(omega_0>0)
sol(t,omega_0,x_0,v_0)=desolve(deq,[x,t],ics=[0,x_0,v_0])
##show(sol)
#x_0=39.24+15=54.24 cm = 0.5424m
#omega_0^2=g/l
val_x_0=0.3924+0.15
val_omega_0=sqrt(9.81/0.3924)
d_sol(t)=sol(t,val_omega_0,val_x_0,0)
show("ecuatia:",d_sol)
show(plot([d_sol(t)],(t,0,10)))
show("Amplitudinea:",val_x_0)
show("faza:",val_omega_0*t)
show("perioada:",2*pi/val_omega_0)
```
