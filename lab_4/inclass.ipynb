```python
#Exercitiul 3
#a)
var("t,r,x_0,r_0,K")
xmalthus=function("xmalthus")(t)
xverhulst=function("xverhulst")(t)
eq1=diff(xmalthus,t,1)==r*xmalthus
sol1=desolve(eq1,[xmalthus,t],ics=[0,x_0])
xmalthus(t,x_0,r)=sol1
show(xmalthus)
eq2=diff(xverhulst,t,1)==r_0*xverhulst*(1-xverhulst/K)
sol2=desolve(eq2,[xverhulst,t],ics=[0,x_0])
sol2=sol2.simplify_log()
xverhulst(t,x_0,r_0,K)=solve(sol2,xverhulst)[0].rhs()
show(xverhulst)
```
